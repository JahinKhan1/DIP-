```python
import cv2
import matplotlib.pyplot as plt

# Step 2: Read an RGB image
input_image = cv2.imread('man.jpg')  # Replace with the path to your image

# Step 3: Create a negative image using cv2.bitwise_not()
negative_image = cv2.bitwise_not(input_image)

# Concatenate the original and negative images horizontally
combined_image = cv2.hconcat([input_image, negative_image])

# Display the combined image using matplotlib
plt.imshow(cv2.cvtColor(combined_image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
from PIL import Image
from IPython.display import display

# Step 1: Open the image using PIL
input_image = Image.open('man.jpg')  # Replace with the path to your image

# Step 2: Create a negative image by inverting pixel values
width, height = input_image.size
negative_image = Image.new("RGB", (width, height))

for x in range(width):
    for y in range(height):
        r, g, b = input_image.getpixel((x, y))
        inverted_pixel = (255 - r, 255 - g, 255 - b)
        negative_image.putpixel((x, y), inverted_pixel)

# Step 3: Show the negative image in the notebook
display(negative_image)
```

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np  # Don't forget to import NumPy

# Step 1: Open the image using PIL
input_image = Image.open('man.jpg')  # Replace with the path to your image

# Step 2: Create a negative image by inverting pixel values
width, height = input_image.size
negative_image = Image.new("RGB", (width, height))

for x in range(width):
    for y in range(height):
        r, g, b = input_image.getpixel((x, y))
        inverted_pixel = (255 - r, 255 - g, 255 - b)
        negative_image.putpixel((x, y), inverted_pixel)

# Convert PIL Image to OpenCV format for concatenation
input_image_cv = cv2.cvtColor(np.array(input_image), cv2.COLOR_RGB2BGR)
negative_image_cv = cv2.cvtColor(np.array(negative_image), cv2.COLOR_RGB2BGR)

# Step 3: Display the original and negative images side by side
combined_image = cv2.hconcat([input_image_cv, negative_image_cv])

# Display the combined image using matplotlib
plt.imshow(cv2.cvtColor(combined_image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np  # Don't forget to import NumPy

# Step 1: Open the image using PIL
input_image = Image.open('man.jpg')  # Replace with the path to your image

# Step 2: Create a negative image by inverting pixel values
width, height = input_image.size
negative_image = Image.new("RGB", (width, height))

for x in range(width):
    for y in range(height):
        r, g, b = input_image.getpixel((x, y))
        inverted_pixel = (255 - r, 255 - g, 255 - b)
        negative_image.putpixel((x, y), inverted_pixel)

plt.subplot(121),plt.imshow(input_image)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(negative_image)
plt.title('Negative Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
