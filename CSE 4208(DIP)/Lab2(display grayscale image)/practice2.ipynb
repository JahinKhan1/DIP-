```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
imgpath = "gray1.jpg"
imgpath1 = "gray2.jpg"
imgpath2 = "gray3.jpg"
imgpath3 = "gray4.jpg"
img = cv2.imread(imgpath,0)
img1 = cv2.imread(imgpath1,0)
img2 = cv2.imread(imgpath2,0)
img3 = cv2.imread(imgpath3,0)
```

```python
plt.subplot(1,4,1)
plt.imshow(img,cmap= 'gray')
plt.title('gray1')
plt.xticks([])
plt.yticks([])
plt.subplot(1,4,2)
plt.imshow(img1,cmap= 'gray')
plt.title('gray2')
plt.xticks([])
plt.yticks([])
plt.subplot(1,4,3)
plt.imshow(img2,cmap= 'gray')
plt.title('gray3')
plt.xticks([])
plt.yticks([])
plt.subplot(1,4,4)
plt.imshow(img3,cmap= 'gray')
plt.title('gray4')
plt.xticks([])
plt.yticks([])
```
