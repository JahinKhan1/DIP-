```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
imgpath = "gray1.jpg"
img = cv2.imread(imgpath,0)
```

```python
plt.subplot(1,2,1)
plt.imshow(img,cmap= 'gray')
plt.title('gray1')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
plt.subplot(1,2,2)
plt.hist(img.ravel(),256,[0,255])
plt.title('Histogram')
plt.show()
```
