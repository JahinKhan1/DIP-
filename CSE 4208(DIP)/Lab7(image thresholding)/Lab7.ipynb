```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

# Load an image from file
image_path = 'lion.jpg'  # Replace with the path to your image file
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded successfully
if original_image is None:
    print("Error: Unable to load the image.")
    exit()

# Set threshold values
thresh_value = 128
max_value = 255

# Apply binary thresholding
ret, binary_threshold = cv2.threshold(original_image, thresh_value, max_value, cv2.THRESH_BINARY)

# Apply inverse binary thresholding
ret, inverse_binary_threshold = cv2.threshold(original_image, thresh_value, max_value, cv2.THRESH_BINARY_INV)

# Display the original and thresholded images
plt.subplot(1, 3, 1), plt.imshow(original_image, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(1, 3, 2), plt.imshow(binary_threshold, cmap='gray')
plt.title('Binary Threshold'), plt.xticks([]), plt.yticks([])

plt.subplot(1, 3, 3), plt.imshow(inverse_binary_threshold, cmap='gray')
plt.title('    Inverse Binary Threshold'), plt.xticks([]), plt.yticks([])

plt.show()
```
