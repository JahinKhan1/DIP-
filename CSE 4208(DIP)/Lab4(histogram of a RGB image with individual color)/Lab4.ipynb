```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
imgpath="RGB.jpg"
img=cv2.imread(imgpath,0)
plt.hist(img)
plt.show()
```

```python
img = cv2.imread('RGB.jpg', cv2.IMREAD_COLOR)
rd, gn, bl = cv2.split(img)
plt.hist(rd.ravel(),bins=256, color='red')
plt.hist(gn.ravel(),bins=256, color='green')
plt.hist(bl.ravel(),bins=256, color='blue')
plt.show()
```

```python
red_c_stretch = cv2.equalizeHist(rd)
green_c_stretch = cv2.equalizeHist(gn)
blue_c_stretch = cv2.equalizeHist(bl)
```

```python
img_stretch = cv2.merge((red_c_stretch,green_c_stretch,blue_c_stretch))
```

```python
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Real")
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img_stretch,cv2.COLOR_BGR2RGB))
plt.title("Stretch")
plt.show()
```
